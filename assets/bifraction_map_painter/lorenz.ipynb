```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_0_330.csv', header=None)
df.fillna(-1.0, inplace=True)
```

```python
# 每行第一个值为 rho，此后为一些点 z
# 以 rho 为横坐标，z 为纵坐标，画出每行的点
# 分岔图
plt.figure(figsize=(20, 20))
for i in range(1, df.shape[1]):
    plt.plot(df.iloc[:, 0], df.iloc[:, i], 'o', markersize=0.05, color='k', alpha=1)
```
